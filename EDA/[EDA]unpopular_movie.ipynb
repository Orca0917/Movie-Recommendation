```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def get_unpopular_item(rating_df):
    df = rating_df.pivot_table(index=['item'], aggfunc='size').sort_values(ascending=True)
    #df = df[df < 200]
    #ret = df.index.to_numpy()
    return df

rating_df = pd.read_csv("/opt/ml/input/data/train/rating.csv")
df = get_unpopular_item(rating_df)
```

```python
fig, axes = plt.subplots(1,2, figsize = (30,6))
idx = list(range(0, len(df)))

axes[0].bar(idx,df)
df2 = df[df < 200]
idx2 = list(range(0, len(df2)))

axes[1].bar(idx2,df2)
plt.ylim(0,500)
plt.xlim(0)

plt.xlabel("items")
plt.ylabel("count")
plt.show()
```
